```python
import cv2
import matplotlib.pyplot as plt 
import IPython.display
#%matplotlib inline


cap = cv2.VideoCapture(1)  # Or try 1, 2, etc.
if not cap.isOpened():
    print("Error: Unable to open the camera.")
else:
    ret, frame = cap.read()
    if not ret:
        print("Error: Unable to read frame from camera.")
    else:
        print("Frame read successfully.")
        # ピクセルごとの色の配列をjpgフォーマットのバイト列に変換
        
        frame = cv2.flip(frame,1)
        _, bytesImage = cv2.imencode('.jpg', frame)

        # jpgフォーマットのバイト列をIPython.display.Imageオブジェクト化
        image = IPython.display.Image(bytesImage)

        # 表示
        IPython.display.display(image)

        
        # Process frame here
    cap.release()
```
